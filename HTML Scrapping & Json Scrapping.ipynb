```python
# Perform json scrapping on the json file Employeedata.json given in Datasets folder and create a structured data comprising only id,name,email and creditBalance
```

```python
import pandas as pd
# import urllib,json
# url="https://jsonplaceholder.typicode.com/users"
# response=urllib.request.urlopen(url)
# data=json.loads(response.read())
data = pd.read_json('../EmployeeData.json')
data.head()
iD=[]
name=[]
email=[]
creditBalance=[]
for index, item in data.iterrows():
    if 'id' in item.keys():
        iD.append(item['id'])
    else:
        iD.append("NA")
    if 'name' in item.keys():
        name.append(item['name'])
    else:
        name.append('NA')
    if 'email' in item.keys():
        email.append(item['email'])
    else:
        email.append('NA')
    if 'creditBalance' in item.keys():
        creditBalance.append(item['creditBalance'])
    else:
        creditBalance.append('NA')
user_df=pd.DataFrame()
user_df['ID']=iD
user_df['NAME']=name
user_df['EMAIL']=email
user_df['CREDIT BALANCE']=creditBalance
user_df.head()
```

```python
# Write a python program to perform html Scrapping on any web url
```

```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
url="https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area"
page=urlopen(url)
htmlpage=page.read().decode("utf-8")
soup=BeautifulSoup(htmlpage,"html.parser")
table=soup.find("table")
SrNo=[]
Country=[]
Area=[]
rows=table.find("tbody").find_all("tr")
# print(rows)
for row in rows:
    cells=row.find_all("td")
    if(cells):
        SrNo.append(cells[0].get_text().strip("\n"))
        Country.append(cells[1].get_text().strip("\xa0").strip("\n"))
        Area.append(cells[3].get_text().strip("\n").replace(",",""))
# print(SrNo)
# print(Country)
# print(Area)
country_frame=pd.DataFrame()
country_frame["ID"]=SrNo
country_frame["Country"]=Country
country_frame["Area"]=Area
print(country_frame.head(20))
```
