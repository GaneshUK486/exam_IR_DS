```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
data=pd.read_csv("Abalone_data.csv")
print("First 5 Rows of Data Frame is : \n",data.head())
print("\n\nINFO of DataFrame is : \n",data.info())
data_x=pd.DataFrame(data.iloc[:,1:8])
data_y=pd.DataFrame(data.iloc[:,-1])
print("The data_x is: \n",data_x.head())
print("The data_y is: \n",data_y.head())
X_train,X_test,Y_train,Y_test = train_test_split(data_x,data_y,test_size=0.5)
reg_model=LinearRegression()
reg_model.fit(X_train,Y_train)
Y_pred=reg_model.predict(X_test)
print(Y_pred)
Y_pred_df=pd.DataFrame(Y_pred,columns=["Predicted"])
print(Y_pred_df)
plt.scatter(Y_test,Y_pred,c="green")
plt.xlabel("Actual Rings")
plt.ylabel("Predicted Rings")
plt.title("Actual Rings vs Predicted Rings")
plt.show()
```
